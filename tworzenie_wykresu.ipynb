Loading data in two columns format

```python
x =[]
y = []
for line in open("S25DH_43_MOD_2-md1_md1_60_rmsd.csv"):
    res = line.split(",")
    x.append(float(res[0]))
    y.append(float(res[1]))
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.xlabel('Time [ps]')
plt.ylabel('Distance [Ã…]')
plt.savefig("S25DH_43_MOD_2-md1_md1_60_rmsd.svg")
plt.show()
```
