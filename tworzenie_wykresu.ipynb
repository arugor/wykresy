Loading data in two columns format

```python
x =[]
y = []
for line in open("reference.csv"):
    res = line.split(",")
    x.append(float(res[0]))
    y.append(float(res[1]))
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.savefig("plot.svg")
plt.show()
```
